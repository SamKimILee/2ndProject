```python
text1 = df.iloc[0:12866].copy()
text1.tail
```

```python
stop_words = pd.read_csv('불용어/2_stopword_fixed.txt', header=None)

def remove_stopwords(text, stop_words):
    # 불용어 제거
    for stop_word in stop_words[0].values:
        text = text.replace(stop_word, '')
    return text

# 불용어 제거 적용
text1['Sentence_cleaned'] = text1['Sentence'].apply(lambda x: remove_stopwords(x, stop_words))

text1.head(50)
```

```python
from konlpy.tag import Mecab
m = Mecab(dicpath="C:/mecab/mecab-ko-dic")
```

```python
tagset = pd.DataFrame(list(m.tagset.items()), columns=['태그', '품사'])
tagset.to_excel('tagset.xlsx', index=False)
```

```python
text2 = text1.drop('Sentence', axis=1)
text2['Sentence_cleaned'] = text2['Sentence_cleaned'].apply(lambda x: m.pos(x))
text2
```

```python
tags = ['NNG', 'NNP', 'VA', 'VV', 'XR']
# 'NNG': 일반 명사
# 'NNP': 고유 명사
# 'VA': 형용사
# 'VV': 동사
# 'XR': 어근
```

```python
text2['Sentence_cleaned'] = text2['Sentence_cleaned'].apply(lambda x: [i[0] for i in x if i[1] in tags])
text2
```

```python
stop_words = ['하', '보']
text2['Sentence_cleaned'] = text2['Sentence_cleaned'].apply(lambda x: [i for i in x if  i not in stop_words])
text2
```

```python
text2 = text2[text2['Sentence_cleaned'].apply(lambda x: len(x) != 0)]
text2
```
