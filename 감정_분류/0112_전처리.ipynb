```python
import pandas as pd
```

```python
text1 = df.iloc[0:50].copy()
text1
```

```python
stop_words = pd.read_csv('불용어/2_stopword_fixed.txt', header=None)

def remove_stopwords(text, stop_words):
    # 불용어 제거
    for stop_word in stop_words[0].values:
        text = text.replace(stop_word, '')
    return text

# 불용어 제거 적용
text1['Sentence_cleaned'] = text1['Sentence'].apply(lambda x: remove_stopwords(x, stop_words))

text1.head(50)
```

```python
from konlpy.tag import Mecab
m = Mecab(dicpath="C:/mecab/mecab-ko-dic")
```

```python
text2 = text1.drop('Sentence', axis=1)
text2['Sentence_cleaned'] = text2['Sentence_cleaned'].apply(lambda x: m.morphs(x))
text2
```
