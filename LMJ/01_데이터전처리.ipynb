```python
import numpy as np
import pandas as pd
import MeCab
import os
import re # 정규표현식

# 환경 변수로 mecab 경로 설정
os.environ["MECABRC"] = "C:/mecab/mecabrc"
```

##### 1. 데이터 불러오기

```python
sentences = pd.read_excel('C:/labs_python/src_mj/datasets/communication_data.xlsx')
sentences_emotion = sentences[['Sentence', 'Emotion']]
```

##### 1-1. 데이터 전처리

```python
sentences_emotion['strip_sentence'] = sentences_emotion['Sentence'].apply(lambda x: x.strip() if isinstance(x, str) else x)

sentences_emotion
```

```python
def clean_korean_text(text):
    # 정규식으로 한글, 숫자, 공백만 남기고 나머지는 제거
    return re.sub(r'[^\uAC00-\uD7A3a-zA-Z0-9\s]', '', text)
```

```python
sentences_emotion['clean_sentence'] = sentences_emotion['strip_sentence'].apply(lambda x: clean_korean_text(x))
```

```python
# 데이터 확인
sentences_emotion
```

##### 1-2 hanspell 라이브러리로 맞춤법 교정

```python
from hanspell import spell_checker
```

```python
# 교정할 문장
# text = "안녕하세요. 저는 자연어 처리를 배우고 있읍니다."

# 맞춤법 검사
# result = spell_checker.check(text)

# 결과 출력
# print("원문:", result.original)  # 입력된 원문
# print("교정된 문장:", result.checked)  # 교정된 결과
# print("교정 여부:", result.result)  # 교정 여부 (True/False)
# print("오류 개수:", result.errors)  # 교정된 오류 수
```

```python
# 맞춤법 교정 함수
def korean_checker(text):
    result = spell_checker.check(text)
    return result.checked
```

```python
sentences_emotion['spell_check_sentence'] = sentences_emotion['clean_sentence'].apply(korean_checker)

sentences_emotion
```

```python
# csv 파일로 저장(all)
output_file = 'C:/labs_python/src_mj/datasets/all_sentences.csv'
sentences_emotion.to_csv(output_file, index=False, encoding='utf-8')

print(f"sentences_emotion을 {output_file} 경로에 저장 완료료.")
```

```python
# csv 파일로 저장(문장 앞 뒤 공백 제거&문장 부호 제거&맞춤법 검사 결과 + emotion)
output_file = 'C:/labs_python/src_mj/datasets/spell_checked_sentences.csv'
sentences_emotion[['Emotion', 'spell_check_sentence']].to_csv(output_file, index=False, encoding='utf-8')

print(f"sentences_emotion을 {output_file} 경로에 저장 완료료.")
```
